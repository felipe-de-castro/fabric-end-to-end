```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%%sql

create table if not EXISTS project_lakehouse.gold_date
(
    Business_date date,
    Business_year int,
    Business_month int,
    Business_quarter int,
    Business_week int
)
using DELTA
PARTITIONED by (Business_year)
```

```python
start_date = '2015-01-01'
end_date = '2030-12-31'

date_diff = spark.sql("select date_diff('{}','{}')".format(end_date,start_date)).first()[0]
```

```python
date_diff
```

```python
ID_Data = spark.range(0,date_diff)
```

```python
date_data = ID_Data.selectExpr(" date_add('{}', cast(id as int)) as Business_date ".format(start_date))
```

```python
date_data.createOrReplaceTempView("ViewDate")
```

```python
%%sql

--delete from project_lakehouse.gold_date;

insert into project_lakehouse.gold_date
select 
Business_date,
Year(Business_date) as Business_year,
Month(Business_date) as Business_month,
quarter(Business_date) as Business_quarter,
weekday(Business_date) as Business_week
from ViewDate;
```
